Build a logistic regression model using TF2.0 to classify points based on a binary label

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Generate synthetic binary classification data
x=np.linspace(-1,10,200).reshape(-1,1)
y=(1/(1+np.exp(-0.8*x))).reshape(-1)>0.5
y=y.astype(np.float32)
```

```python
#Create logistic regression model
model=Sequential([
    Dense(units=1,activation='sigmoid',input_shape=[1])
])
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=100,verbose=0)
```

```python
probs=model.predict(x)
preds=(probs>0.5).astype(np.float32)
```

```python
plt.scatter(x,y,label='True Labels',alpha=0.5)
plt.scatter(x,probs,color='red',label='Prediction Probabilities')
plt.title('Logistic Regression with Tensorflow 2')
plt.legend()
plt.show()
```
