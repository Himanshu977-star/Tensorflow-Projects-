Build a simple linear regression model using TF2.0 to learn relationship y=3x+2 for synthetic data

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Generate synthetic training data
x=np.linspace(0,10,100)
y=3*x+2+np.random.randn(*x.shape)*0.5
```

```python
model=Sequential([
    Dense(units=1, input_shape=[1])
])
```

```python
model.compile(optimizer='sgd',loss='mse')
```

```python
model.fit(x,y,epochs=100,verbose=0)
```

```python
predictions=model.predict(x)
```

```python
plt.scatter(x,y,label='Data')
plt.plot(x,predictions,color='red',label='Model')
plt.legend()
plt.title('Linear Regression with Tensorflow 2')
plt.show()
```
